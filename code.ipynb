```python
import pandas as pd
import numpy as np
from sklearn.linear_model import LinearRegression
import matplotlib.pyplot as plt
from sklearn.linear_model import RANSACRegressor
```

```python
dff=df[(df["district"]=="جیحون")&(df["shot"]==2023)].groupby("year")["price_per_meter"].mean()
np.array(dff.index).reshape(-1, 1)
```

```python
try:    
    dist=df['district'].unique()
    df['Depreciation_rate'] = np.nan 
    for reg in dist:
        # ایجاد مدل رگرسیون خطی
        model = LinearRegression()
        dff=df[(df["district"]==reg)&(df["shot"]==2023)].groupby("year")["price_per_meter"].median()#.plot()
        dff=dff[dff.index>=1390]
        # آموزش مدل با داده‌های X و y
        X=np.array(dff.index).reshape(-1, 1)
        model.fit(X, dff.values)

        # استخراج شیب و عرض از مبدا (ضرایب خط)
        slope = model.coef_  # شیب خط
        intercept = model.intercept_  # عرض از مبدا
        # ابتدا ستون جدید را با NaN پر می‌کنیم
        df.loc[df["district"]==reg, 'Depreciation_rate'] = slope[0]  # فقط ردیف‌هایی که شرط برقرار است را تغییر می‌دهیم
        if reg=="حسن‌آباد":
            print(dff)
            # نمایش نتایج
            print(f"y = {slope[0]}x + {intercept}")
            # پیش‌بینی مقادیر برای خط رگرسیون
            y_pred = model.predict(X)
            print(reg,"\n")
            # رسم داده‌ها و خط رگرسیون
            plt.scatter(X, dff.values, color='blue', label='Data Points')
            plt.plot(X, y_pred, color='red', label='Regression Line')
            plt.show()
except:
    print(dff,reg,df[(df["district"]==reg)&(df["shot"]==2023)].groupby("year")["price_per_meter"].mean())
    df.loc[df["district"]==reg, 'Depreciation_rate'] = float(df[df['district']=="پونک"]["Depreciation_rate"].head(1).values)
```

```python
def adjust_price(row):
    age = row['age']  # فرض بر این است که ستون 'building_age' سن ساختمان را نشان می‌دهد
    price_per_meter = row['price_per_meter']

    if age < 2:
        return price_per_meter  # اگر سن بنا کمتر از 2 سال بود، قیمت تغییر نمی‌کند
    elif 2 <= age <= 10:
        discount_factor = 1 - ((age) * 0.015)  # برای هر سال بین 2 تا 6، نیم درصد تخفیف
        return price_per_meter / discount_factor
    else:  # اگر سن بنا بیشتر از 6 سال بود
        discount_factor = 1 - ((10) * 0.025)  # برای هر سال بالاتر از 6، یک درصد تخفیف
        return price_per_meter / discount_factor

# ایجاد ستون جدید با اعمال تابع adjust_price
df["unreal_price_per_meter"] = df.apply(adjust_price, axis=1)
df[df["district"]=="شهرک گلها"]
```
